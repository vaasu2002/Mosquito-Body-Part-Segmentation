```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.io
import skimage.feature
import skimage.color
import skimage.transform
import skimage.util
import skimage.segmentation
import numpy
from skimage.segmentation import mark_boundaries
```

```python
plt.figure(figsize=(20,20))
```

```python
img_path = "data/source_images/mosquitoes/4115_back.jpg"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
im_mask = skimage.segmentation.felzenszwalb(
        skimage.util.img_as_float(img), scale= 150, sigma= 1,
        min_size= 75)
bound_img = mark_boundaries(img, im_mask)
plt.figure(figsize=(20,20))
plt.subplot(231)
plt.imshow(img)
plt.subplot(232)
plt.imshow(im_mask)
plt.subplot(233)
plt.imshow(bound_img)
plt.show()
```
